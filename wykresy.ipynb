```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
params = {'font.size': 16,
          'figure.figsize': (12.,8.),
          'axes.labelsize': 16,
          'xtick.labelsize': 16,
          'ytick.labelsize': 16
         }
plt.rcParams.update(params)

palette = ['#1b9e77', '#d95f02', '#7570b3'] # https://colorbrewer2.org
palette2 = ['#762a83', '#9970ab', '#c2a5cf', '#a6dba0', '#5aae61', '#1b7837']
```

```python
dane_karzel = pd.read_csv('envtab_karzel_v2.dat', comment='#', sep="\s+")
dane_rgb = pd.read_csv('envtab_rgb_v2.dat', comment='#', sep="\s+")
dane_agb = pd.read_csv('envtab_agb_v2.dat', comment='#', sep="\s+")
```

```python
plt.figure(1,tight_layout=True)

plt.plot(dane_karzel['r'], dane_karzel['rho'], lw=3, c=palette[0], label='karzeÅ‚ MS')
plt.plot(dane_rgb['r'], dane_rgb['rho'], lw=3, c=palette[1], label='olbrzym RGB')
plt.plot(dane_agb['r'], dane_agb['rho'], lw=3, c=palette[2], label='olbrzym AGB')

plt.xlabel('promieÅ„ (R$_\odot$)')
plt.ylabel('gÄ™stoÅ›Ä‡ (g/cm$^3$)')

plt.xscale('log')
plt.yscale('log')
plt.legend()

plt.savefig('promien-gestosc.pdf', bbox_inches='tight')
```

```python
plt.figure(2,tight_layout=True)

plt.plot(dane_karzel['r'], dane_karzel['t'], lw=3, c=palette[0], label='karzeÅ‚ MS')
plt.plot(dane_rgb['r'], dane_rgb['t'], lw=3, c=palette[1], label='olbrzym RGB')
plt.plot(dane_agb['r'], dane_agb['t'], lw=3, c=palette[2], label='olbrzym AGB')

plt.xlabel('promieÅ„ (R$_\odot$)')
plt.ylabel('temperatura (K)')

plt.xscale('log')
plt.yscale('log')
plt.legend()

plt.savefig('promien-temperatura.pdf', bbox_inches='tight')
```

```python
plt.figure(3,tight_layout=True)

plt.plot(dane_karzel['r'], dane_karzel['vt'], lw=3, c=palette[0], label='karzeÅ‚ MS')
plt.plot(dane_rgb['r'], dane_rgb['vt'], lw=3, c=palette[1], label='olbrzym RGB')
plt.plot(dane_agb['r'], dane_agb['vt'], lw=3, c=palette[2], label='olbrzym AGB')

plt.xlabel('promieÅ„ (R$_\odot$)')
plt.ylabel('prÄ™dkoÅ›Ä‡ konwekcji (cm/s)')

plt.xscale('log')
plt.yscale('log')
plt.legend(loc='upper left')

plt.savefig('promien-predkosc.pdf', bbox_inches='tight')
```

```python
fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(12,12))
plt.subplots_adjust(left=0.08, right=0.96, bottom=0.05, top=0.98)

ax[0].plot(dane_karzel['r'], dane_karzel['rho'], lw=3, c=palette[0], label='karzeÅ‚ MS')
ax[0].plot(dane_rgb['r'], dane_rgb['rho'], lw=3, c=palette[1], label='olbrzym RGB')
ax[0].plot(dane_agb['r'], dane_agb['rho'], lw=3, c=palette[2], label='olbrzym AGB')

ax[1].plot(dane_karzel['r'], dane_karzel['t'], lw=3, c=palette[0], label='karzeÅ‚ MS')
ax[1].plot(dane_rgb['r'], dane_rgb['t'], lw=3, c=palette[1], label='olbrzym RGB')
ax[1].plot(dane_agb['r'], dane_agb['t'], lw=3, c=palette[2], label='olbrzym AGB')

ax[2].plot(dane_karzel['r'], dane_karzel['vt'], lw=3, c=palette[0], label='karzeÅ‚ MS')
ax[2].plot(dane_rgb['r'], dane_rgb['vt'], lw=3, c=palette[1], label='olbrzym RGB')
ax[2].plot(dane_agb['r'], dane_agb['vt'], lw=3, c=palette[2], label='olbrzym AGB')

for i in range(2):
    ax[i].axes.xaxis.set_visible(False)

ax[2].set_xlabel('promieÅ„ (R$_\odot$)')
ax[0].set_ylabel('gÄ™stoÅ›Ä‡ (g/cm$^3$)')
ax[1].set_ylabel('temperatura (K)')
ax[2].set_ylabel('prÄ™dkoÅ›Ä‡ konwekcji (cm/s)')

for i in range(3):
    ax[i].set_xscale('log')
    ax[i].set_yscale('log')

ax[0].legend()
fig.align_ylabels(ax[:])
plt.subplots_adjust(hspace=0)

plt.savefig('promien.pdf', bbox_inches='tight')
```
